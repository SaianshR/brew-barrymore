```python
df_sim_scores['name_review'] = df_sim_scores['product_name'] + " " + df_sim_scores['product_review']
df_sim_scores
```

```python
beer_attributes = ['chocolate', 'dark', 'sweet', 'bourbon']

all_beer_names = [i for i in np.unique(df_sim_scores['product_name'])]
rand_beers = [all_beer_names[np.random.randint(0,len(all_beer_names))] for i in range(0, 9, 1)]

beer_attributes_names = [i for i in beer_attributes]
for i in rand_beers:
    beer_attributes_names.append(i)

col = 'name_review'
lift_table, lift_dict_values = ra.lift_ratio(index=beer_attributes_names, data=df_sim_scores, capture_column=col)
```

```python
lift_dict_values
```

```python
rando_beer = lift_dict_values.columns[np.random.randint(0, len(lift_dict_values))]
rando_beer
```

```python
lift_dict_values.columns
```

```python
diff = []
idx = []
for i in lift_dict_values.columns[4:]:
    if i == rando_beer:
        continue
    series = abs(lift_dict_values[rando_beer][:4]) - abs(lift_dict_values[i][:4])
    val = sum(series.values)
    diff.append(val)
    idx.append(i)
```

```python
most_sim = pd.Series(diff, index=idx)
most_sim
```

For our code, we randomly select 10 beers and then randomly select 1 beer from those ten. This random beer is compared to others through a taking the lift values across the 4 attributes (in our case: "chocolate," "dark," "sweet," "bourbon") and summing the whole series of values for each beer (we now have 10 beers with a 1 value that represents the sum of their lift values on their attributes). We then take the difference between the combined lift for the selected beer and the remaining beers. The closer the remaining values are to zero, the more similar that one of the nine beers is the selected beer. During the code's testing, we had the following 10 beers:
